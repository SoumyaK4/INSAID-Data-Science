**<h4>Question 1:** Create a function that generates frequency distribution of categorical features.</h4>

---

<details>

**<summary>Hint:</summary>**

- Create a figure size of 15 X 7 inches.

- Extract labels and values using .value_counts() function.

- Use pandas .plot.bar() function to create a bar chart.

- Set xticks rotation to zero degrees.

- Set x, y labels of size 14 and title label of size 16.

- Display the output.

</details>

```python
def plotBarChart(data=None, column=None, xlabel=None):
  """Creates bar chart depicting frequency distribution of categorical features."""

  # Create a figure size of 15 X 7 inches.
  figure = plt.figure(figsize=[15, 7])

  # Using pandas .plot.bar() function for barplot
  data[column].value_counts().plot.bar()

  # Changning x-ticks to 0 degree
  plt.xticks(rotation=0)

  # Set x, y labels of size 14 and title label of size 16.
  plt.xlabel(xlabel=xlabel, size=14)
  plt.ylabel(ylabel='Frequency', size=14)
  plt.title(label='Frequency Distribution of '+column, size=16)

  # Display the output
  plt.show()
```

```python
plotBarChart(data=data, column='Type 1', xlabel='Type 1')
```

**<h4>Question 2:** Create a function that generates area distribution of numerical features.</h4>

---

<details>

**<summary>Hint:</summary>**

- Create a figure size of 15 X 7 inches.

- Use pandas .plot.area() function to create area chart.

- Set x, y labels of size 14 and title label of size 16.

- Display the output.

</details>

```python
def plotAreaChart(data=None, column=None, xlabel=None):
  """Creates area chart depicting area distribution of numerical features."""
  
  # Create a figure size of 15 X 7 inches.
  figure = plt.figure(figsize=[15, 7])

  # Using pandas .plot.area()
  data[column].value_counts().sort_index().plot.area()

  # Set x, y labels of size 14 and title label of size 16.
  plt.xlabel(xlabel=xlabel, size=14)
  plt.ylabel(ylabel='Value', size=14)
  plt.title(label='Area Distribution of '+column, size=16)

  # Display the output
  plt.show()
```

```python
plotAreaChart(data=data, column='Total', xlabel='Total')
```

```python
def plotHistChart(data=None, column=None, bins=10, xlabel=None):
  """Creates histogram chart depicting frequency distribution of numerical features."""

  # Create a figure size of 15 X 7 inches.
  figure = plt.figure(figsize=[15, 7])

  # Using .hist() function to pass the labels and values.
  data[column].plot.hist(bins=bins)

  # Set x, y labels of size 14 and title label of size 16.
  plt.xlabel(xlabel=xlabel, size=14)
  plt.ylabel(ylabel='Frequency', size=14)
  plt.title(label='Frequency Distribution of '+column, size=16)

  # Display the output
  plt.show()
```

**<h4>Question 5:** Create a function that generates a line plot of numerical features.</h4>

---

<details>

**<summary>Hint:</summary>**

```python
def plotLineChart(data=None, xaxis=None, yaxis=None):
  """Creates scatter chart of two numerical features."""

  # Sort the data by xaxis in ascending order
  data = data.sort_values(by=xaxis)

  # Using pandas .plot.line() function
  data.plot.line(x=xaxis, y=yaxis, figsize=[15, 7])

  # Set x, y labels of size 14 and title label of size 16.
  plt.xlabel(xlabel=xaxis, size=14)
  plt.ylabel(ylabel=yaxis, size=14)
  plt.title(label='Line Distribution of '+xaxis+' & '+yaxis, size=16)

  # Display the output
  plt.show()
```

**<h4>Question 6:** Create a function that generates a pie plot only if number of categories are less than 7.</h4>

---

<details>

**<summary>Hint:</summary>**

- Check if the number of categories are less than 7 or not.

- If the number of categories are less than 7 then do as following.

  - Create a figure size of 15 X 7 inches.

  - Create explode list of length equal to the number of categories using np.ones() and divide by 10.

  - Use pandas .pie() function and pass the data column along with additional parameters.

  - Display the percentage of the categories starting at angle 180 and wedge proportion of width of 0.15.

  - Set title label of size 16 and display the output.

- If the number of categories are greaer than equal to 7 then display error message.

</details>

```python
def plotPieChart(data=None, column=None):
  """Creates a pie chart if categories are less than 7."""

  # Check if number of categories are less than 7
  if len(data[column].unique()) < 7:

    # Create a figure size of 15 X 7 inches.
    figure = plt.figure(figsize=[15, 7])

    # Initiate an explode list of categories
    explode_list = np.ones(len(data[column].unique()))/10

    # Using pandas .plot.pie() function
    data[column].value_counts().plot(kind='pie', explode=explode_list, 
                                     autopct='%3.1f%%', wedgeprops=dict(width=0.15), 
                                     shadow=True, startangle=160, 
                                     cmap='inferno', legend=True)
    
    plt.axis('equal')

    # Set title label of size 16.
    plt.title(label='Pie Distribution of '+column, size=16)

    # Display the output
    plt.show()

  else:
    print('Error! Number of categories are greater than equal to 7.')
```

```python
plotPieChart(data=data, column='Legendary')
```

**<h4>Question 7:** Create a function that generates a box plot of numerical features.</h4>

---

<details>

**<summary>Hint:</summary>**

- Create a figure size of 15 X 7 inches.

- Use pandas .boxplot() function to create a box plot.

- Set x, y labels of size 14 and title label of size 16.

- Display the output.

</details>

```python
def plotBoxChart(data=None, column=None):
  """Creates box chart of numerical features."""

  # Create a figure size of 15 X 7 inches.
  figure = plt.figure(figsize=[15, 7])

  # Using pandas .boxplot() function
  data.boxplot(column=column)

  # Set x, y labels of size 14 and title label of size 16.
  plt.xlabel(xlabel=column, size=14)
  plt.ylabel(ylabel='Values', size=14)
  plt.title(label='Box Distribution of '+column, size=16)

  # Display the output
  plt.show()
```

```python
plotBoxChart(data=data, column='Total')
```

**<h4>Question 8:** Create a function that generates a hexbin plot of numerical features.</h4>

---

<details>

**<summary>Hint:</summary>**

- Use pandas .plot.hexbin() function to create a hexbin plot.

- You can pass a figure size of 15 X 7 inches.

- Add a colorbar using .colorbar() function to read the graph.

- Set x, y labels of size 14 and title label of size 16.

- Display the output.

</details>

```python
def plotHexbinChart(data=None, xaxis=None, yaxis=None):
  """Creates box chart of numerical features."""

  # Using pandas .plot.hexbin() function
  data.plot.hexbin(x=xaxis, y=yaxis, color='#4CB391', figsize=[15, 7], gridsize=25, sharex=False)

  # Set x, y labels of size 14 and title label of size 16.
  plt.xlabel(xlabel=xaxis, size=14)
  plt.ylabel(ylabel=yaxis, size=14)
  plt.title(label='Hexbin Distribution of '+xaxis+' & '+yaxis, size=16)

  # Display the output
  plt.show()
```

```python
plotHexbinChart(data=data, xaxis='Attack', yaxis='Defense')
```
