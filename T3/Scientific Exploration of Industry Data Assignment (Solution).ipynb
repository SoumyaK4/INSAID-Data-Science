---
**<h4>Question 1:** Create a function that replaces the null values with the appropriate values.</h4>

---

<details>

**<summary>Hint:</summary>**

- Use .isna().sum() to check for number of null values in the dataset.

- Replace continuous features' missing values with the mean of that feature.

- For categorical features, use mode of that feature to fill in missing values.

- Use .fillna() method to fill in the null values in the dataset. 

</details>

```python
def fill_null_values(data=None):

  # culmen_depth_mm is a continuous feature, we can the replace null values with mean of the feature.
  data['culmen_depth_mm'] = data['culmen_depth_mm'].fillna(value=data['culmen_depth_mm'].mean())

  # culmen_length_mm is a continuous feature, we can the replace null values with mean of the feature.
  data['culmen_length_mm'] = data['culmen_length_mm'].fillna(value=data['culmen_length_mm'].mean())

  # flipper_length_mm is a continuous feature, we can the replace null values with mean of the feature.
  data['flipper_length_mm'] = data['flipper_length_mm'].fillna(value=data['flipper_length_mm'].mean())

  # body_mass_g is a continuous feature, we can the replace null values with mean of the feature.
  data['body_mass_g'] = data['body_mass_g'].fillna(value=data['body_mass_g'].mean())

  # sex is a categorical feature, we can the replace null values with mode of the feature.
  data['sex'] = data['sex'].fillna(value=data['sex'].mode()[0])
  
  return data.isna().sum()
```

```python
fill_null_values(data=data)
```

- We have **successfully cleaned** our data.

**<h4>Question 2:** Create a function that displays the frequency & proportion of each species of penguins in the dataset.</h4>

---

<details>

**<summary>Hint:</summary>**

- You can use a countplot to plot the count for each species.

- Set the figure size as 7 x 7 inches

- Set x, y labels of size 14, ticks of size 12 and title label of size 16.

- Beautify the plot to make it look appealing.

</details>

```python
def species_counter(data=None):
  
  # Create a figure of size 7x7
  fig = plt.figure(figsize=(7, 7))

  # Plot the countplot using sns.countplot and species feature.
  ax = sns.countplot(x='species', data=data, palette='rocket')

  # Adding percentages to the bars
  for p in ax.patches:
    total = data.shape[0]
    percentage = '{:.2f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 5
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y))

  # Set x, y labels of size 14, ticks of size 12 and title label of size 16.
  plt.xlabel(xlabel='Species', size=14)
  plt.ylabel(ylabel='Count', size=14)
  plt.xticks(size=12)
  plt.yticks(size=12)
  plt.title(label='Species Count', size=16)

  #Put a grid
  plt.grid(b=True)

  # Display the plot
  plt.show()
```

```python
species_counter(data=data)
```

**Observations**:

- **Adelie** penguin species are **more than twice** in number as **Chinstrap** penguins.

**<h4>Question 4:** Create a function that displays the frequency of each penguin species across all the islands in the dataset.</h4>

---

<details>

**<summary>Hint:</summary>**

- You can use the .countplot() method to plot the frequency for each species across all the islands.

- Keep x parameter as island

- Set the figure size as 7 x 7 inches

```python
def species_island_counter(data=None):
  
  # Create a figure of size 7x7
  fig = plt.figure(figsize=(7, 7))

  # Plot the countplot using sns.countplot with x=island and y=species.
  ax = sns.countplot(x='island', hue='species', data=data, palette='rocket')
  
  # Adding percentages to the bars
  for p in ax.patches:
    total = data.shape[0]
    percentage = '{:.2f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 10
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y))

  # Set x, y labels of size 14, ticks of size 12 and title label of size 16.
  plt.xlabel(xlabel='Islands', size=14)
  plt.ylabel(ylabel='Count', size=14)
  plt.xticks(size=12)
  plt.yticks(size=12)
  plt.title(label='Species Count across islands', size=16)

  #Put a grid
  plt.grid(b=True)

  # Display the plot
  plt.show()
```
