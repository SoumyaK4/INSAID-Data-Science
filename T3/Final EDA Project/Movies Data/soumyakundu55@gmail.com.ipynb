### **6.1 Changing column names**

```python
movies_df.columns
```

```python
movies_df.rename(columns = {'Runtime (Minutes)':'Runtime_Mins', 'Revenue (Millions)': 'Revenue_Mils'}, inplace = True)
```

```python
movies_df.info()
```

### **6.2 Handling of Missing Data & '0' Values**

- In this section, we will take appropriate measures for missing data.

**Previous Observations:**

- We can observe that following features are found to have missing values along with the proportions:

|Feature|Data Type|Missing Proportion|Solution|
|:--:|:--:|:--:|:--|
|Revenue_Mils|float|12.8%|Replace with median.|
|Metascore|Float|6.4%|Replace with median.|

- **Revenue_Mils has '0' values**, drop the datapoints.

```python
movies_df['Revenue_Mils'].fillna(movies_df['Revenue_Mils'].median(), axis=0, inplace=True)
movies_df['Metascore'].fillna(movies_df['Metascore'].median(), axis=0, inplace=True)
movies_df = movies_df[movies_df['Revenue_Mils']!=0]
```

```python
null_frame = pd.DataFrame(index = movies_df.columns.values)
null_frame['Null Frequency'] = movies_df.isnull().sum().values
null_percent = movies_df.isnull().sum().values / movies_df.shape[0]
null_frame['Missing %'] = np.round(null_percent, decimals = 4)*100
null_frame.transpose()
```

```python
movies_df.Revenue_Mils.min()
```

```python
movies_df.drop("Description", axis=1, inplace=True)
```

```python
movies_df[["Metascore"]] = movies_df[["Metascore"]].astype("int")
```

## Q1: What is the number of movies released year wise?

```python
fig = plt.figure(figsize=(15,8))                                          

sns.countplot(x='Year', data=movies_df, palette = 'magma')                                         

plt.title(label='No. of Movies released per Year', size=22)
plt.xlabel(xlabel='Year', size=18)
plt.ylabel(ylabel='Frequency', size=16)
plt.xticks(size=10)
plt.yticks(size=10)
plt.grid(visible=True)

plt.show()
```

#### **Observations**

```python
figure = plt.figure(figsize=[10, 10])

space = np.ones(11)/10

movies_df['Year'].value_counts().plot(kind='pie', explode=space, fontsize=14, 
                                       autopct='%3.1f%%', wedgeprops=dict(width=0.15), 
                                       shadow=True, startangle=100, cmap='inferno')

plt.ylabel('')
plt.title(label='Proportion of Movies released each year', size=18)
plt.show()
```
